```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the data
file_path = 'MY DATA MURAT.xlsx'  # Specify the file path
df = pd.read_excel(file_path)

# Select only numeric columns
numeric_columns = df.select_dtypes(include=['number'])

# Basic statistical summary
print("Basic Statistical Features:")
print(numeric_columns.describe())
```

```
# Correlation matrix
correlation_matrix = numeric_columns.corr()

# Graphical summary (histograms, boxplot, and correlation heatmap)
plt.figure(figsize=(20, 12))

# 1. Histograms
plt.subplot(2, 2, 1)
numeric_columns.hist(ax=plt.gca(), bins=15, color='skyblue')
plt.title("Histograms")

plt.tight_layout()
plt.show()
```

```
import matplotlib.pyplot as plt

# Scatter plot for Calories vs. Exercise Duration
plt.figure(figsize=(10, 6))
plt.scatter(df['Calories Intake (kcal)'], df['Exercise Duration (min)'], alpha=0.7, color='g')
plt.title('Calories Intake vs. Exercise Duration', fontsize=14)
plt.xlabel('Calories Intake (kcal)', fontsize=12)
plt.ylabel('Exercise Duration (min)', fontsize=12)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```
import matplotlib.pyplot as plt

# Create a figure and two axes (subplots)
fig, ax1 = plt.subplots(figsize=(12, 6))

# Plot Weight over Time
ax1.plot(df['Date'], df['Weight (kg)'], color='blue', label='Weight (kg)', marker='o')
ax1.set_xlabel('Date', fontsize=12)
ax1.set_ylabel('Weight (kg)', color='blue', fontsize=12)
ax1.tick_params(axis='y', labelcolor='blue')
ax1.set_title('Impact of Exercise on Weight Over Time', fontsize=14)
ax1.grid(True, linestyle='--', alpha=0.5)

# Add a second y-axis for Exercise Duration
ax2 = ax1.twinx()
ax2.scatter(df['Date'], df['Exercise Duration (min)'], color='green', label='Exercise Duration (min)', alpha=0.7)
ax2.set_ylabel('Exercise Duration (min)', color='green', fontsize=12)
ax2.tick_params(axis='y', labelcolor='green')

# Combine the legends
fig.legend(loc='upper left', bbox_to_anchor=(0.1, 0.9), fontsize=12)

plt.tight_layout()
plt.show()
```

```
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

# Load the data
file_path = 'MY DATA MURAT.xlsx'  # Specify the file path
df = pd.read_excel(file_path)

# Convert the 'Date' column to datetime format and set it as the index
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Column to forecast (e.g., 'Steps' column)
target_column = 'Steps'

# Select the time series
time_series = df[target_column]

# Define the ARIMA model (initial values: p=1, d=1, q=1)
model = ARIMA(time_series, order=(1, 1, 1))
model_fit = model.fit()

# Forecast for future dates (e.g., predict the next 10 days)
forecast_steps = 10  # Number of days to forecast
forecast = model_fit.forecast(steps=forecast_steps)

# Print the forecast results
print(f"Forecasted values for the next {forecast_steps} days:")
print(forecast)

# Visualize the forecast results (optional)
plt.figure(figsize=(10, 6))
plt.plot(time_series, label='Actual Values')
plt.plot(forecast.index, forecast, label='Forecasted Values', color='red')
plt.title(f"Forecast for {target_column}")
plt.xlabel("Date")
plt.ylabel(target_column)
plt.legend()
plt.grid(True)
plt.show()
```

```
import pandas as pd
from statsmodels.tsa.arima.model import ARIMA
import matplotlib.pyplot as plt

# Load the data
file_path = 'MY DATA MURAT.xlsx'  # Specify the file path
df = pd.read_excel(file_path)

# Convert the 'Date' column to datetime format and set it as the index
df['Date'] = pd.to_datetime(df['Date'])
df.set_index('Date', inplace=True)

# Column to forecast: 'Weight (kg)'
target_column = 'Weight (kg)'
time_series = df[target_column]

# Define and fit the ARIMA model
model = ARIMA(time_series, order=(1, 1, 1))
model_fit = model.fit()

# Forecast the next 30 days
forecast_steps = 30  # Forecast 30 days ahead
forecast = model_fit.forecast(steps=forecast_steps)

# Print the forecast results
print(f"Forecasted weight values for the next {forecast_steps} days:")
print(forecast)

# Visualize the forecast results
plt.figure(figsize=(10, 6))
plt.plot(time_series, label='Actual Weight Values')
plt.plot(forecast.index, forecast, label='Forecasted Weight', color='red')
plt.title("Weight Forecast (Next 30 Days)")
plt.xlabel("Date")
plt.ylabel("Weight (kg)")
plt.legend()
plt.grid(True)
plt.show()
```
